## Functions

```python
def multivariate_lognormal_cascade(n, sigma1=1, sigma2=1, corr=0):
    rng = np.random.default_rng()
    mu1 = -1/2 * sigma1**2
    mu2 = -1/2 * sigma2**2

    PQ = np.exp(rng.multivariate_normal(np.array([mu1,mu2]),  np.array([[sigma1**2,corr*sigma1*sigma2], [corr*sigma1*sigma2, sigma2**2]]), 4))

    P = PQ[:,0].reshape(2,2)
    Q = PQ[:,1].reshape(2,2)
    for i in range(n):
        PQ = np.exp(rng.multivariate_normal(np.array([mu1,mu2]),np.array([[sigma1**2,corr*sigma1*sigma2], [corr*sigma1*sigma2, sigma2**2]]), P.shape[0]**2 * 4))
        P = np.kron(P, np.ones((2,2)))
        P = P * PQ[:,0].reshape(P.shape)
        Q = np.kron(Q, np.ones((2,2)))
        Q = Q * PQ[:,1].reshape(P.shape)

    P = P / np.sum(P)
    Q = Q / np.sum(Q)
    return(np.stack([P,Q], axis=-1))

def theil(pop):
  tp = pop.sum(axis=2)
  prob = pop[:,:,:] / tp[:,:, np.newaxis]
  ep = np.sum(-np.log2(prob) * prob, axis=2)
  e = (ep * tp / tp.sum()).sum()
  return(1 - e)

def scale_theil(pop, scales):
    stheils = []
    for s in scales:
      small_arr = block_reduce(pop, block_size=(s,s,1), func=np.sum)
      stheils.append(theil(small_arr))
    return(stheils)

def theil_hat(pop):
    return((np.sum((pop[:,:,0] + pop[:,:,1])/4 * ( (1/8 * ((pop[:,:,1]/pop[:,:,0])-1)**2) #-
                                         #(1/8 * ((pop[:,:,1]/pop[:,:,0])-1)**3)  + 
                                         #(19/192 *   ((pop[:,:,1]/pop[:,:,0])-1)**4)  -
                                         #(7/96 *   ((pop[:,:,1]/pop[:,:,0])-1)**5) + 
                                         #(101/1920 *((pop[:,:,1]/pop[:,:,0])-1)**6) -
                                         #(73/1920 *((pop[:,:,1]/pop[:,:,0])-1)**7)  +
                                         #(5993/215040 *((pop[:,:,1]/pop[:,:,0])-1)**8) -
                                         #(1121/53760 *((pop[:,:,1]/pop[:,:,0])-1)**9) +
                                         #(10303/645120 *((pop[:,:,1]/pop[:,:,0])-1)**10)-
                                         #(8081/645120  *((pop[:,:,1]/pop[:,:,0])-1)**11)+  
                                         #(570499/56770560  *((pop[:,:,1]/pop[:,:,0])-1)**12)+         
                                                  
                                         +(1/8 * ((pop[:,:,0]/pop[:,:,1])-1)**2) 
                                         #-(1/8 * ((pop[:,:,0]/pop[:,:,1])-1)**3)  + 
                                         #(19/192 *   ((pop[:,:,0]/pop[:,:,1])-1)**4)  -
                                         #(7/96 *   ((pop[:,:,0]/pop[:,:,1])-1)**5) + 
                                         #(101/1920 *((pop[:,:,0]/pop[:,:,1])-1)**6) -
                                         #(73/1920 *((pop[:,:,0]/pop[:,:,1])-1)**7)  +
                                         #(5993/215040 *((pop[:,:,0]/pop[:,:,1])-1)**8) -
                                         #(1121/53760 *((pop[:,:,0]/pop[:,:,1])-1)**9) +
                                         #(10303/645120 *((pop[:,:,0]/pop[:,:,1])-1)**10)-
                                         #(8081/645120  *((pop[:,:,0]/pop[:,:,1])-1)**11)+
                                         #(570499/56770560  *((pop[:,:,0]/pop[:,:,1])-1)**12)         
                                                 
                                                 
                                                 
                                                 
                                                 )     
                                               
                  
                  
                  
                  
                  )) / np.log(2))

def scale_theils(pop, scales):
    stheils = []
    stheil_hat = []
    for s in scales:
      small_arr = block_reduce(pop, block_size=(s,s,1), func=np.sum)
      stheils.append(theil(small_arr))
      stheil_hat.append(theil_hat(small_arr))
    return([stheils, stheil_hat])
```

```python
test1 = multivariate_lognormal_cascade(6, sigma1=0.2**0.5, sigma2=0.2**0.5, corr=.6)
test2 = multivariate_lognormal_cascade(6, sigma1=0.2**0.5, sigma2=0.2**0.5, corr=-.6)

scales = 2**np.array([i for i in range(0,7)])
Duncans1 = []
Duncans2 = []
for s in scales:
  small_arr1 = block_reduce(test1, block_size=(s,s,1), func=np.sum)
  Duncans1.append(Duncan(small_arr1))
  small_arr2 = block_reduce(test2, block_size=(s,s,1), func=np.sum)
  Duncans2.append(Duncan(small_arr2))

plt.figure(figsize=(10, 6))
plt.scatter(scales, Duncans1, label='cln 1')
plt.plot(scales, Duncans1, label='cln 1')
plt.scatter(scales, Duncans2, label='cln 2')
plt.plot(scales, Duncans2, label='cln 2')

plt.xlabel('Scales')
plt.ylabel('Duncan Index')
plt.title('Duncan Index at Different Scales')
plt.legend()
plt.grid(True)
plt.show()
```
